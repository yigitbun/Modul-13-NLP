```python
from HanTa import HanoverTagger as h

tagger = h.HanoverTagger('morphmodel_ger.pgz')

lemmacounter=Counter()

import nltk
import codecs
import glob
from collections import Counter

datei=codecs.open('Actelion.txt','r','utf8')

text=datei.read()
sentences=nltk.sent_tokenize(text,language='german')

for sent in sentences:
    words=nltk.word_tokenize(sent,language='german')
    lemmata=tagger.tag_sent(words)
    lemmacounter.update([lemma for wort,lemma,pos in lemmata if pos=='NN' or pos=='NE'])

lemmacounter.most_common()
```
