```python
df=pd.read_csv("ReviewDetails.csv")
df
```

```python
raitings=[]
for i in df['average_rating'] : 
    raitings.append(int(i.split()[0].split(',')[0])-3)
    
    
```

```python
print(raitings[0])
```

```python
sum(raitings)
```

```python
comments=[]
for i in df['review_text'] : 
    comments.append(i)
```

```python
print(comments[0])
```

```python
count=0
for text in comments:
    blob = TextBlobDE(text) 
    if (blob.sentiment.polarity>0):
        count+= 1
print(count, 'von',len(comments),'positive')
```

```python
for i in range(len(comments)):
    blob = TextBlobDE(comments[i]) 
    if (blob.sentiment.polarity>0):
        print('positive', raitings[i])
```

```python
gleich=0
unterschiedlich=0
for i in range(len(comments)):
    blob = TextBlobDE(comments[i]) 
    if (blob.sentiment.polarity>0):
        if(raitings[i]==1):
            gleich+=1
        else:
            unterschiedlich+=1
print('gleich: ', gleich)
print('unterschiedlich: ', unterschiedlich)
```
