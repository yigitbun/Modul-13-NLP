```python
import pandas as pd
import textblob
from textblob_de import TextBlobDE
```

```python
df = pd.read_csv('ReviewDetails.csv')
```

```python
df.head()
```

```python
df_reviews = df[['review_text','review_rating']]
```

```python
df_reviews.head()
```

```python
df_negativ=df_reviews[df_reviews["review_rating"] == "1,0 von 5 Sternen"]
```

```python
df_negativ.head()
```

```python
df_positiv=df_reviews[df_reviews["review_rating"] == "5,0 von 5 Sternen"]
```

```python
df_positiv.head()
```

```python
nKommentar = []
for n in df_negativ['review_text']:
    nKommentar.append(n)
```

```python
print(nKommentar[2])
```

```python
len(nKommentar)
```

```python
for k in nKommentar:
    blob = TextBlobDE(k)
    if blob.sentiment.polarity<0.5:
        print('Kommentar', ' neg')
    else:
        print('falsch')
```

```python
#count = 0
for k in nKommentar:
    blob = TextBlobDE(k)
    if blob.sentiment.polarity<0.5:
        #count+=1
        print('Kommentar ist: ',' -> neg')
    else:
        print('Kommentar ist: ', ' -> falsch bewertet')
```

```python
pKommentar = []
for n in df_positiv['review_text']:
    pKommentar.append(n)
```

```python
count = 0
for i in pKommentar:
    blob = TextBlobDE(i)
    if blob.sentiment.polarity>-0.1:
        count+=1
        print('Kommentar ist: ',' -> pos')
    else:
        print('Kommentar ist: ', ' -> falsch bewertet')
```
