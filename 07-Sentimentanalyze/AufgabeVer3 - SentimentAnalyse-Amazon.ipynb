```python
import textblob
from textblob_de import TextBlobDE
```

```python
import pandas as pd
```

```python
df=pd.read_csv("ReviewDetails.csv")
df
```

```python
raitings=[int(i.split()[0].split(',')[0]) for i in df['review_rating']]
comments=[i for i in df['review_text']]
```

```python
print(raitings[0])
print(comments[0])
```

```python
for i in range(len(comments)):
    blob = TextBlobDE(comments[i]) 
    print(blob.sentiment.polarity,'------>',raitings[i])
```

```python
gleich=0
unterschiedlich=0
for i in range(len(comments)):
    blob = TextBlobDE(comments[i]) 
    if (blob.sentiment.polarity>0):
        if(raitings[i]>2.5):
            gleich+=1
        else:
            unterschiedlich+=1
    elif (blob.sentiment.polarity<=0):
        if(raitings[i]<2.5):
            gleich+=1
        else:
            unterschiedlich+=1
    
    
print('gleich: ', gleich)
print('unterschiedlich: ', unterschiedlich)
```
